```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv('wiki_movie_plots_deduped.csv')
# Use 'on_bad_lines' instead of 'error_bad_lines'
# 'skip' will skip bad lines, you can also use 'warn' to get a warning.
df.info()
df.head()
```

```python
print(df['Cast'].isnull().sum())

print(df[df['Cast'].isnull()])
print(df[df['Cast'].isnull()]['Genre'].value_counts())

print(df[df['Cast'].isnull()]['Director'].value_counts())
```

```python
df = df.dropna(subset=['Cast'])

print(df['Cast'].isnull().sum())
df.info()
```

```python
df = df.drop(columns=['Wiki Page'])
```

```python
def combine_columns(row):
    combined_string = ""
    for column in df.columns:
        if column != 'combined':
            value = str(row[column])
            combined_string += f"{column}: {value}, "
    return combined_string[:-2]

df['combined'] = df.apply(combine_columns, axis=1)

df.info()
df.head()
```

```python
df['ID'] = range(1, len(df) + 1)
df = df[['ID', 'Title', 'combined']]
df.head()
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
import pickle
```

```python

model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(df["combined"].tolist(), show_progress_bar=True)
embeddings = np.array(embeddings, dtype=np.float32)

index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(embeddings)

faiss.write_index(index, "movie_faiss.index")

```

```python

with open("movie_data.pkl", "wb") as f:
    pickle.dump(df[["ID", "Title", "combined"]], f)

print("✅ FAISS Index & Metadata Saved")
```

```python

index = faiss.read_index("movie_faiss.index")

with open("movie_data.pkl", "rb") as f:
    movie_data = pickle.load(f)

query = "A movie where many heros band together to defeat an intergalactic enemy"

query_embedding = model.encode([query], convert_to_numpy=True)

k = 5
distances, indices = index.search(query_embedding, k)

recommended_movies = movie_data.iloc[indices[0]]

print("\nTop Recommended Movies:")
for _, row in recommended_movies.iterrows():
    print(f"{row['Title']} - {row['combined']}\n")
```
